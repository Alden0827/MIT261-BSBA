```python
### my updated working code snippets:

# Create a list of subjects the student has passed
passed_subjects = student_grades_df[student_grades_df['Grade'] >= 75]['Subject Code'].tolist()

# Filter for subjects for the current semester and not yet taken
potential_subjects = curriculum_df[
    (curriculum_df['semester'] == 'Second') &
    (curriculum_df['year'] == 2) &
    (~curriculum_df['Subject Code'].isin(student_grades_df["Subject Code"].tolist()))
].copy()

available_subjects = []
blocked_subjects = []

for _, row in potential_subjects.iterrows():
    prerequisites = row['preRequisites']
    
    # Check if the prerequisites list is empty or if all prerequisites are in the passed_subjects list
    if not prerequisites or all(prereq in passed_subjects for prereq in prerequisites):
        available_subjects.append(row)
    else:
        blocked_subjects.append(row)

# Convert the lists of rows back to DataFrames
available_subjects_df = pd.DataFrame(available_subjects)
blocked_subjects_df = pd.DataFrame(blocked_subjects)

```
